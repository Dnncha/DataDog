```python
import matplotlib.pyplot
from get_data import *

def closing_lastn(n):
    """Closing price for the last n days"""
    data=get_data()
    dlen=len(data) #length of data (days)
    cp=data.Close[dlen-n:dlen] # closing price last n days
    return cp
   
n=365*2
cp=closing_lastn(n)
cp_mean=mean(cp) # mean price last n days
cp_med =median(cp)
print "mean price last", n, "days:", cp_mean
print "median price last", n, "days:", cp_med
print cp.plot(figsize=(15, 10))
```
